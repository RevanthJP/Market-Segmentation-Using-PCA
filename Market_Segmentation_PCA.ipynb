# 4) Check the dataset for outliers before and after scaling. Draw your inferences from this exercise.

- As, we know that only the Ecom, SalesFImage, OrdBilling and DelSpeed has outliers from the previous data exploration done using unscaled data.
- Let check for the same variables for outliers in the scaled data.

```python
plt.figure(figsize = (20,7))
salon_scaled.boxplot();
```

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3 = col.quantile([0.25,0.75])
    IQR = Q3 - Q1  
    lower_range = Q1 -(1.5*IQR)
    upper_range = Q3 +(1.5*IQR)
    return lower_range,upper_range
```

```python
remove_outlier(salon_scaled['Ecom'])
```

```python
salon_scaled['Ecom'] = np.where(salon_scaled['Ecom']> 1.7618, salon_scaled['Ecom'].median(), salon_scaled['Ecom'])
salon_scaled['Ecom'] = np.where(salon_scaled['Ecom']< -1.9684, salon_scaled['Ecom'].median(), salon_scaled['Ecom'])
```

# 6) Write the explicit form of the first PC (in terms of Eigen Vectors)

Before we proceed further, we need to check the significance of correlation and adequacy of the features using bartlett sphericity test and kmo.

#### Barlett sphericity test
- H0 : All the features are significantly correlated.
- H1 : There is no significant correlationn


```python
from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity

chi_square_value, p_value = calculate_bartlett_sphericity(salon_scaled)
print(p_value)
```

Here, we can reject null hypothesis as p-value is less than 0.05.

#### kmo
- In kmo, if measure of sampling adequacy(MSA) is less than 0.5, PCA is not recommended, since no reduction is expected. On the other hand, MSA > 0.7 is expected to provide a considerable reduction in the dimension and extraction of meaningful components.

```python
from factor_analyzer.factor_analyzer import calculate_kmo

kmo_all, kmo_model = calculate_kmo(salon_scaled)
print(kmo_model)
```

Since MSA is close to 0.7, we can expect a considerable reduction in the dimension and extraction of meaningful components to some extent.

#### PCA

```python
len(salon_scaled.columns)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components= 8, random_state=1)
salon_pca = pca.fit_transform(salon_scaled)
salon_pca.T
```

```python
pca.components_
```

```python
pca.explained_variance_
```

```python
var_exp = pca.explained_variance_ratio_
var_exp
```

Observation shows that the first 5 components explains the variance better.

```python
plt.figure(figsize=(12,7))
sns.lineplot(y=var_exp,x=range(1,len(var_exp)+1),marker='o')
plt.xlabel('Number of Components',fontsize=15)
plt.ylabel('Variance Explained',fontsize=15)
plt.title('Scree Plot',fontsize=15)
plt.grid()
plt.show()
```

```python
pca_salon = pd.DataFrame(pca2.components_, columns=salon_scaled.columns)
pca_salon.head()
```

```python
from matplotlib.patches import Rectangle

fig,ax = plt.subplots(figsize=(22, 10), facecolor='w', edgecolor='k')
ax = sns.heatmap(pca_salon, annot=True, vmax=1.0, vmin=0, cmap='Blues', cbar=False, fmt='.2g', ax=ax,
                 yticklabels=['PC0','PC1','PC2','PC3','PC4'])

column_max = pca_salon.abs().idxmax(axis=0)

for col, variable in enumerate(pca_salon.columns):
    position = pca_salon.index.get_loc(column_max[variable])
    ax.add_patch(Rectangle((col, position),1,1, fill=False, edgecolor='red', lw=3))
```
